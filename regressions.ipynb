```python
import numpy as np

links = np.array([1, 2, 3, 4, 5])
end_result = np.array([0, 0, 0, 1, 1])

w = 0.0 
b = 0.0

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict_probability(links, w, b):
    z = w*links+b
    return sigmoid(z)

def loss_function(end_result, model_prediction):
    return -np.mean(end_result * np.log(model_prediction + 1e-9)+ (1-end_result)+np.log(1-model_prediction+1e-9))

performing_rate = 0.1
echoes = 2000

for i in range(echoes):
    model_prediction = predict_probability(links, w, b)
    dw = np.mean((model_prediction - end_result)*links)
    db = np.mean(model_prediction-end_result)

    w = w - performing_rate * dw
    b = b - performing_rate* db

    if i % 500 == 0:
        loss = loss_function(end_result, model_prediction)
        print("Echoes", i, "loss", loss)
def predict(links, w, b):
    probs = predict_probability(links, w, b)
    return (probs >= 0.5).astype(int)

email_links = 3

prob = predict_probability(email_links, w, b)
result = predict(email_links,w, b)

print("Probability of spam", prob)
print("Result", "spam" if result==1 else "not spam")

        
```
